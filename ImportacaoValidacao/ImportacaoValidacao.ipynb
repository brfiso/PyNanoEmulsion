# Importação dos Lipídios

```python
df_informacoes = pd.read_csv('LipidiosCompletos.csv', sep=';')
df_informacoes.head()
```

```python
# Remover colunas Unnamed
colunas_para_remover = [col for col in df_informacoes.columns if "Unnamed" in col]
df_informacoes.drop(columns=colunas_para_remover, inplace=True)
```

```python
from statistics import mean


def aplicar_media(intervalo:str):    
    valores = str(intervalo).split('|')
    media = mean([float(valor.replace(',', '.')) for valor in valores])

    return media


for col in df_informacoes.columns:
    if col == 'Lipidio':
        continue 
    
    df_informacoes[col] = pd.to_numeric(df_informacoes[col].apply(aplicar_media))

df_informacoes.to_csv('lipidios-finais.csv', sep=';', encoding='utf-8-sig')
```
